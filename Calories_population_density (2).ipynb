Data sourced from wikipedia

```python
# import libraries

import pandas as pd
```

```python
data1 = pd.read_html('https://en.wikipedia.org/wiki/List_of_countries_by_food_energy_intake')
data2 = pd.read_html('https://simple.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population_density')
```

```python

# assign calories table to variable
calories = data1[1]
calories
```

```python
# assign population density table to variable
density = data2[0]
density
```

```python
# rename calories columns
calories.columns = ['rank', 'country', 'calories', 'joules']
calories
```

```python
# rename density coloumns
density.columns = ['rank', 'country', 'areakm','areami','population','densitykm','densitymi','date','source']
density
```

```python
# drop calories columns
calories.drop(calories.filter(regex='rank').columns, axis=1, inplace=True)
calories
```

```python
# drop columns in density using column index
cols = [0,2,5,7,8]
density.drop(density.columns[cols],axis=1,inplace=True)
density
```

```python
# merge calories and density tables

table = pd.merge(calories, density, on=['country'], how = 'left')
table
```

```python
# Saving the result
table.to_csv('FAO_master_population.csv', encoding='latin1')
```
